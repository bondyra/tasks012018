## Data description

Delivered data contains 5043 observations from imdb movie database. Each observation, a movie, is described by 27 attributes, regarding names of main actors, film directors and many more. 

## Observations

This section contains facts and dependencies, that I have observed and explored in the data. My main areas of interests concerned:
- simple aggregates, showing most popular actors and directors in data
- basic correlations of numerical attributes (like duration and budget)
- analysis of keywords and movie genre, also with association rules
- giving answers to some interesting questions, related to data and concerning categorical and numerical attributes dependencies
- unsupervised dimensionality reduction based on PCA
- clustering the data with k-means algorithm
- setting regression problems on data (prediction of movie score and movie revenue) and solving them with popular regression models

### Numbers in data

Firstly, I have explored the basic counts of movies regarding some specific value of categorical attribute. I have noticed following characteristics:
- the most frequent movie director is Steven Spielberg with total count of 26 movies, followed by Woody Allen,
- the most frequent first actor in a movie is Robert De Niro (49 movies),
- the most frequent second actor is Morgan Freeman (20 movies).

I have also checked joint conditions of actors and directors.This way, I have noticed several properties:
- Suprisingly, the first two most frequent pairs (first actor - director) are one person - Woody Allen and Clint Eastwood,
- Summing number of occurences of specific first and sector actor, the most frequent pair is Steve Buscemi and Adam Sandler.

In this stage, I have noticed that budgets of movie are not uniform - some of the movies have budget described in different than USD currency - as I have checked by addressing links of movies provided in data. This problem could cause problems with further data exploration, so I had to remove some of the observations by using simple  thresholding (if budget was absurdly high, I removed observation).

### Correlations in data

I have checked the correlations (Pearson and Spearman) of numerical attributes in data. I concluded following observations:
- Highly correlated attributes (>0.5 for absolute value of correlation) are mainly basic social statistics, like facebook likes or number of reviews. Many of these attributes are strictly linearly dependent
- Some of the correlated attributes vaguely describe abstract concept of movie popularity. For example, attributes like gross (revenue) and number of user reviews describes overall movie popularity, that's why they are correlated.

Some of most correlated pairs are shown below:

Very visible attribute, that is nearly not correlated with anything, is attribute describing number of faces in movie posters.

### Movie genres and movie keywords

I have checked genres and keywords characteristics of movies. These attributes were completely denormalized, so proper regex parsing had to be perfomed.

The most popular are drama and comedy movies.
Genres occurence in data can be presented by simple histogram:

### Hypotheses testing

I tried to answer to some interesting questions related directly to data:
- Is director somehow correlated with budget/gross?

To answer this question, I have analysed budgets and revenues (gross) of movies directed by most popular directors (to get reasonable amount of data for each person). Apparently, there is not significant difference between budget and revenues of many of these directors, except for Steven Spielberg. For movies of this director, revenue is larger than budget. For formal justification of my reasoning, I have used Wilcoxon signed-rank tests for not normally distributed data of highly correlated pairs.

- Is actor somehow correlated to budget/gross?

I have performed similar operations to these above. I haven't noticed any statistically significant differences between movie budget and revenue for most popular actors.

- Is rating or origin country affecting budget/gross?

Answer is - the best revenues are from movies rated PG-13 and PG, and from movies that come from US.

- Are there any changes of attributes mean according to year?

There is one interesting time-dependent variable - budget. By reading the graph, we can see drastic drop of mean in budgets of movies in 2008. This is the year when global financial crisis started.

### Segmentation

I have also used k-means clustering algorithms to perform basic segmentation of movies in data.

It is also worh noting that I have used only few attributes of the original dataset. To select proper attributes, I have performed PCA on numerical attributes. These selected attributes, have simply the best loadings for principal components.

Movies can be segmented into five distinct groups, described as:
- **"Creepy" movies** - movies with bad rating and with smallest number of votes (group has 27.3% of movies),
- **Oldies movies** - movies substantially older than movies in other groups with good imdb rating score (9.8% of movies),
- **Average movies** - truly average movies - from small to average budget, average rating (49 % of movies),
- **Must-see movies** - movies are movies with best rating, big number of votes. budget is quite big (4.5% of movies),

```python
# biplot generation taken from: https://github.com/teddyroland/python-biplot/blob/master/biplot.py
def get_biplot (df, pca=None, text=True):
    if pca is None:
        pca = PCA(n_components = 2)
        pca.fit(df)

    xvector = pca.components_[0]
    yvector = pca.components_[1]

    xs = pca.transform(df)[:,0]
    ys = pca.transform(df)[:,1]

    for i in range(len(xs)):
        plt.plot(xs[i], ys[i], 'bo')
        #dont plot text (to make biplot more readable)
        if (text):
            plt.text(xs[i]*1.2, ys[i]*1.2, list(df.index)[i], color='b')
            
    for i in range(len(xvector)):
        plt.arrow(0, 0, xvector[i]*max(xs), yvector[i]*max(ys),
                  color='r', width=0.0005, head_width=0.0025)
        plt.text(xvector[i]*max(xs)*1.2, yvector[i]*max(ys)*1.2,
                 list(df.columns.values)[i], color='r')

    plt.show()
    return pca
```
